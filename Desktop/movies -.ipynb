```python
import pandas as pd 
import numpy as np

df_credit = pd.read_csv('tmdb_5000_credits.csv')
df_movies = pd.read_csv('tmdb_5000_movies.csv')


df_credit.rename(columns={'movie_id' : 'id'},inplace = True)
df = df_movies.merge(df_credit, on = 'id')
df.shape
```

```python
df.columns
```

```python
df = df[['id', 'genres', 'title_x', 'overview' , 'cast' , 'crew', 'keywords']]
df.head()
```

```python
df.columns.tolist()
```

```python
import ast
def convert(obj):
    data = ast.literal_eval(obj)  #CONVERT STRINGS INTO LIST OF DICTS
    names = []
    for item in data:
        names.append(item['name'])  #each time it takes the name and add it to the names list
    return names


df['genres'] = df['genres'].apply(convert)
df['keywords'] = df['keywords'].apply(convert) #to the dataframe
                            
df.head()
```

```python
#CONVERT LIST INTO STRING IN ONE LINE 
```

```python
df['genres'] = df['genres'].apply(lambda x: " ".join(x) if isinstance(x, list) else "")
df['overview'] = df['overview'].apply(lambda x: " ".join(x) if isinstance(x, list) else "")
df['cast'] = df['cast'].apply(lambda x: " ".join(x) if isinstance(x, list) else "")
df['crew'] = df['crew'].apply(lambda x: " ".join(x) if isinstance(x, list) else "")
df['keywords'] = df['keywords'].apply(lambda x: " ".join(x) if isinstance(x, list) else "")
```

```python
# ISINSTANCE = BEFORE JOINING THIS CHECK IF X REALLY A LIST , ELSE RETURN TO EMPKTY STRING
```

```python
#CREATING NEW COLUMN
```

```python
df['tags'] = df['overview'] + " " + df['genres']+ " " +df['cast'] +" " + df['crew']+ " "+df['keywords']

df_final = df[['id', 'title_x', 'tags']]
df_final.rename(columns ={'title_x': 'title'} , inplace = True)
```

```python
# " " TO MAKE SURE WORDS FROM DIFFERENT COLUMNS DONT GET JOINED TOGETHER DIRECTLY. 
```

```python
# COVERT TEXT INTO NUMERIC NUMBERS
```

```python
#CONTENT BASED RECOM
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
#THIS GIVES SIMILARITY SCORE BETWEEN EVERY PAIR OF MOVIES, IF ANGLE IS SMALL, SIMLARITY IS HIGH AND IF ANGLE IS BIG THEN SIMILARITY IS LOW
```

```python
tfidf = TfidfVectorizer(max_features=5000, stop_words='english')
vectors = tfidf.fit_transform(df_final['tags']).toarray()
similarity= cosine_similarity(vectors)
#MAX_FEATURES= 5000: USES THE TOP 5000 WORDS ONLY .
#ENGLISH WORDS LIKE THE A IS ETC  
#O AND 1 CHANGE
```

```python
#DEFINE RECOMMENDATION FUNCTION
```

```python
def recommend(movies):
    movie = movies.lower() #CONVERTS THE INPUT MOVIE NAME TO LOWERCASE TO ENSURE CASE_INSENSITIVE MATCHING WITH THE DATASET
    if movie not in df_final['title'].str.lower().values :  #CHECK IF LOWERCASE VERSION OF THE MOVIE EXIST IN DF_FINAL['TITLE']
        return "movie not found in database."
    index = df_final[df_final['title'].str.lower() == movie].index[0]
    distances= list(enumerate(similarity[index]))
    distances= sorted(distances, reverse=True, key=lambda x: x[1])[1:6]

    print(f"\n movies similar to{movie.title()}':\n")
    
    for i in distances:
       print(df_final.iloc[i[0]].title)
        #CONVERT LOC(TITLE) 
                      
    
```

```python
recommend("Avatar")
```
